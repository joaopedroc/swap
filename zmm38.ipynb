```python
%sql

-- Consulta principal
SELECT 
    ekko.BUKRS,           -- Empresa
    ekko.EBELN,           -- Contrato/Pedido Serviço
    ekko.BSART,           -- Classe de documento
    ekko.STATU,           -- Status
    ekko.LIFNR,           -- Fornecedor
    ekko.BEDAT,           -- Data de criação
    ekko.KDATB,           -- Data início
    ekko.KDATE,           -- Data fim
    -- Dados do item
    ekpo.EBELP,           -- Posição
    ekpo.MATNR,           -- Material
    ekpo.MENGE,           -- Quantidade
    ekpo.NETWR,           -- Valor líquido
    -- Releases
    eket.ETENR,           -- Número da exigência
    eket.MENGE AS MENGE_RELEASE,  -- Quantidade liberada
    eket.EINDT,           -- Data de entrega
    -- Atribuições
    ekkn.AUFNR,           -- Ordem de serviço
    ekkn.KOSTL            -- Centro de custo
FROM EKKO ekko
LEFT JOIN EKPO ekpo ON ekko.EBELN = ekpo.EBELN
LEFT JOIN EKET eket ON ekpo.EBELN = eket.EBELN AND ekpo.EBELP = eket.EBELP
LEFT JOIN EKKN ekkn ON ekpo.EBELN = ekkn.EBELN AND ekpo.EBELP = ekkn.EBELP



-- WHERE ekko.BUKRS = ?      -- Empresa
--   AND ekko.EBELN = ?      -- Contrato/Pedido Serviço
--   AND ekko.BSTYP = 'K'    -- Contratos
```
